Видим большое количество уникальных записей, которые выбиваются из критерия оценки. Проверим процентное соотношение некорректных ответов в выборке, и можем ли мы их отбросить.

```python
# Отбрасываем пустые варианты ответов
data = data[~data.Q1.isna()]

# Списое корректных оценок из теста
mark_list = [str(x) for x in range(1,11)]

# Заменяем значения "0" на "1"
data.loc[data.Q1 == '0', 'Q1'] = '1'

# Вычисляем процент некорретных ответов на Q1
percent = data[~data.Q1.isin(mark_list)]['Q1'].count()/data.Q1.count()*100

print(f"Процент некорректных ответов: {round(percent,2)} %")
```

Процент некорретных ответов незначителен, значит мы можем отбросить их из выборки.

```python
data = data[data.Q1.isin(mark_list)]
data.Q1 = data.Q1.astype('int64')
```

Условимся, что оценка >=7 говорит, что клиент удовлетворен сервисом, <= 4 - что неудовлетворен, 5 и 6 означает нейтральную позицию. И проранжируем оценки согласно этому.

```python
evaluation = {1: 'not satisfied',
              2: 'not satisfied',
              3: 'not satisfied',
              4: 'not satisfied',
              5: 'neutral',
              6: 'neutral',
              7: 'satisfied',
              8: 'satisfied',
              9: 'satisfied',
              10: 'satisfied'}

data['evaluation'] = data['Q1'].map(evaluation)
```

Отрисуем гистограмму распределения оценок на вопрос Q1

```python
def plot_hist(data, col, title=None, xlab=None, ylab=None, median=False, mean=False, hue=None):
    """Функция отрисовки гистограммы"""

    plt.figure(figsize=(8, 4))
    sns.histplot(data, x=col, kde=False, hue=hue, discrete=True, alpha = 0.5)
    plt.title(title, fontsize=14, pad=15)
    plt.ylabel(ylab)
    plt.xlabel(xlab)
    plt.xticks(np.arange(data[col].min(), data[col].max()+1))

    if mean is True:
        plt.axvline(data[col].mean(), color='m',
                    linestyle='dashed', label='среднее', alpha=0.5)
        plt.legend()

    if median is True:
        plt.axvline(data[col].median(), color='green',
                    linestyle='dashed', label='медиана', alpha=0.7)
        plt.legend()

    return plt.show()
```

Посмотрим на наличие зависимостей метрик качества связи.

```python
# Напишем функцию возвращающую значения корреляции между метриками выше установленного фильтра

def data_corr(data, metrics, filter = 0.2):
    data_corr = data.corr()
    corr_list = data_corr[((data_corr > filter) | (data_corr < -filter)) & (data_corr < 1)][metrics].dropna()
    return corr_list
```

Но можно отметить корреляцию "Умеренную" и "Заметную" между параметрами:

* Video Streaming Download Throughput(Kbps);
* Video Streaming xKB Start Delay(ms);
* Web Page Download Throughput(Kbps).

Возможно, если улучшить качество одного из параметров, он позитивно скажется на остальных метриках, что улучшит индекс удовлетворенности клиента в целом.

Для оценки степени влияния технических параметров разобъем датасет на 2 группы:
* 1) удовлетворенных качеством связи клиентов (Q1 > 6), 
* 2) неудовлетворенных качеством связи клиентов (Q1 < 5)

Отобразим гистограммы распределения метрики "Downlink Throughput(Kbps)" для визуального распределения в двух группах. 

```python
def metrics_hist(data, metrics, log=False):    
    """Функция по визуализации причин по которой клиенты снижали оценку.
Сначала убирает нейтральные оценки и сравнивает две группы с хорошими оценками и плохими"""
    # Отсеиваем оценки
    data = data[(data.Q1 >= 7) | (data.Q1 <= 4)]
    # Делим на две группы
    data['evaluation'] = data['Q1']\
        .apply(lambda x: 'Not satisfied' if x <= 4 else 'Satisfied')
    # Формируем полотно
    fig, ax = plt.subplots(figsize=(10, 4))

    fig.suptitle(f'Плотность распределения', y=1., fontsize=14)
    
    sns.histplot(ax=ax, data=data,
             x=metrics, hue='evaluation', stat='density',
             hue_order=data.evaluation.unique(),
             kde=True, log_scale=log, legend=False)
    if log == True:
        ax.set_xlabel('Log: ' + re.findall(r'\((\S+)\)', metrics)[0])
    else:
        ax.set_xlabel(re.findall(r'\((\S+)\)', metrics)[0])
    ax.axvline(np.median(data[data.Q1 > 4][metrics] ), linestyle='--', color='b', linewidth=3)
    ax.axvline(np.median(data[data.Q1 <= 4][metrics] ), linestyle='--', color='r', linewidth=3)
    ax.set_title(metrics)
    plt.legend(data.evaluation.unique())
```

```python
metrics_hist(data, "Downlink Throughput(Kbps)")
```

Также построим гистограмму в логарифмическом масштабе

```python
metrics_hist(data, "Downlink Throughput(Kbps)", log=True)
```

Гистограммы представляют собой ненормальное распределение. Распределение группы неудовлетворенных клиентов имеет большие значения на малых скоростях передачи данных. Однако, судя по мединному значению, кажется, что скорость передачи данных для группы удовлетворенных клиентов в среднем выше. Проверим это утверждение с помощью статистики.

## 1.3 Статистический тест

Для проведения статистического теста определим 2 гипотезы:

H0: Скорость не оказывает влияния на удовлетворенность абонента сотовой связью. \
H1: Скорость передачи данных имеет статистически значимые различия между двумя группами, а именно у большинства удовлетворенных связью клиентов лучше качество связи.

Будем использовать бутстрэп, из-за несбалансированности выборок и недостаточного количества наблюдений. 

```python
def bootstrap(data_a, data_b, N=2000, title=None, conf_level = 95):
    """Функция бутстрэпа, расчет и отрисовка"""
    
    # Реализация бутстрэпа
    mean_samples_a = []
    mean_samples_b = []
    buf = []
    size_a = data_a.shape[0] 
    size_b = data_b.shape[0] 
    
    for i in range(N):
        sample_a = np.random.choice(data_a.values, size_a, True)
        mean_samples_a.append(np.mean(sample_a))
        sample_b =  np.random.choice(data_b.values, size_b, True)
        mean_samples_b.append(np.mean(sample_b))
        buf.append(np.mean(sample_a)-np.mean(sample_b))
    
    buf_cent = np.array(buf)  - np.mean(buf)
    p_value = np.mean(buf_cent >= np.mean(data_a) - np.mean(data_b))
     # Инициализация полотна
    fig, ax = plt.subplots(1, 2, figsize=(16, 6))
    fig.suptitle(title, y=1.05, fontsize=14)
    
    # Графики распределения по группам А и Б
    sns.histplot(mean_samples_a, alpha=.8, ax=ax[0], kde=True, stat='density')
    sns.histplot(mean_samples_b, color='orange', alpha=.8, ax=ax[0], kde=True, stat='density')
    # Интервалы и медиана
    left_quant = (100 - conf_level)/2
    right_quant = 100 - (100 - conf_level) / 2
    
    ax[0].axvline(np.percentile(mean_samples_a, left_quant), linestyle='-', color='r', linewidth=3)
    ax[0].axvline(np.percentile(mean_samples_a, right_quant), linestyle='-', color='r', linewidth=3)
    ax[0].axvline(np.median(mean_samples_a), linestyle='--', color='r', linewidth=3)
    ax[0].axvline(np.percentile(mean_samples_b, left_quant), linestyle='-', color='g', linewidth=3)
    ax[0].axvline(np.percentile(mean_samples_b, right_quant), linestyle='-', color='g', linewidth=3)
    ax[0].axvline(np.median(mean_samples_b), linestyle='--', color='g', linewidth=3)
    ax[0].set_title('Плотность распределения по группам \n удовлетворенных и неудовлетворенных сервисом клиентов', fontsize=14)
    ax[0].set_ylabel('Плотность')
    ax[0].set_xlabel(re.findall(r'\((\S+)\)', title)[0])
    ax[0].legend(['Satisfied','Not satisfied'])
    
    # График распределения разности по группам А и Б
    sns.histplot(buf_cent, alpha=.8, ax=ax[1], kde=True, stat='density', color='g')
    # Интервалы и медиана
    ax[1].axvline(np.percentile(buf_cent, left_quant), linestyle='-', color='b', linewidth=3)
    ax[1].axvline(np.percentile(buf_cent, right_quant), linestyle='-', color='b', linewidth=3)
    ax[1].axvline(np.mean(data_a) - np.mean(data_b), linestyle='--', color='b', linewidth=3)
    ax[1].set_title('Центрированная плотность разницы средних', fontsize=14)
    ax[1].set_ylabel('Плотность')
    ax[1].set_xlabel(re.findall(r'\((\S+)\)', title)[0])
    print("Bootstrap p-value:", p_value)
    return plt.show()
```

В целом данные коррекно, заполнены. Но встречаются пропуски и двузначные числа больше 10. Вероятно пропущены запятые между ответами. Обработаем столбец. Уберем не валидные значения и задвоения.

```python
# Заменяем пустые варианты ответов на значение 6 - то есть затрудняюсь ответить
data_q2.Q2.fillna('6', inplace=True)
```

```python
def q2_map(ans):
    """ Функция по обработке ответов Q2"""
    """ans - строка с ответами на вопрос
       функция возвращает список цифр диапазона ответа без повторов"""

    # Регулярным выражением делим строку на список чисел
    ex = re.findall('\w+', ans)

    # Пробразуем к целому
    ex = list(map(int, ex))

    # Новый список для записи
    new_list = []

    # Итерируемся по списку и проверяем на соответстие условию
    for element in ex:
        if 1 <= element <= 7:
            new_list.append(element)  # Если верный диапазон отметки
        elif element > 10:
            # Если двузначное число, делим по символам и проверяем вхождение в диапазон
            new_list += [int(x) for x in str(element) if 1 <= int(x) <= 7]
        else:
            pass
    # На выходе удаляем дубликаты
    return list(set(new_list))
```

```python
# Разбиение строки в столбце на список значений(ответов) на Q2
data_q2['Q2'] = data_q2['Q2'].apply(lambda x: q2_map(x))
```
